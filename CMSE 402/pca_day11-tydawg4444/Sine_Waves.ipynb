```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
import matplotlib.animation as animation
%matplotlib inline
```

```python
Range = np.linspace(0, 2*np.pi, 100)
```

```python
image_number = 0
for i in range(0,len(Range)):
    x = []
    y = []
    for j in Range[0:i]:
        x.append(j)
        y.append(np.sin(j))
    plt.plot(x, y)
    plt.title("Sine Graph")
    image_number = '{:03d}'.format(image_number)
    plt.savefig("image_"+image_number+".png")
    image_number = int(image_number) + 1
```
